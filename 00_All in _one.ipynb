```python
# ==========================================
# PANDAS FINAL REVISION
# Author : Yash Maroti Petkar
# ==========================================

# 1. Import Pandas
import pandas as pd

print("Pandas Imported Successfully")
print("-" * 50)

# ==========================================
# 2. Read CSV File
# ==========================================

df = pd.read_csv("G:\downloads\student_marks.csv")

print("Dataset Loaded Successfully")
print(df)

print("-" * 50)

# ==========================================
# 3. Head & Tail
# ==========================================

print("First 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())

print("-" * 50)

# ==========================================
# 4. Describe Dataset
# ==========================================

print("Statistical Summary")
print(df.describe())

print("-" * 50)

# ==========================================
# 5. Dataset Information
# ==========================================

print("Dataset Information")
print(df.info())

print("-" * 50)

# ==========================================
# 6. Column Selection
# ==========================================

print("Single Column")
print(df["Maths"])

print("\nMultiple Columns")
print(df[["Physics", "Maths"]])

print("-" * 50)

# ==========================================
# 7. Row Selection
# ==========================================

print("First Row")
print(df.loc[0])

print("\nRows 0 to 2")
print(df.loc[0:2])

print("-" * 50)

# ==========================================
# 8. Fill Null Values
# ==========================================

# Check null values
print("Null Values")
print(df.isnull().sum())

# Fill null values with 0
df.fillna(0, inplace=True)

print("\nNull Values After Filling")
print(df.isnull().sum())

print("-" * 50)

# ==========================================
# 9. Rename Column
# ==========================================

df.rename(columns={"Marks": "Student_Marks"}, inplace=True)

print("Columns After Rename")
print(df.columns)

print("-" * 50)

# ==========================================
# 10. Change Datatype
# ==========================================

df["Student_Marks"] = df["Maths"].astype(float)

print("Datatype Changed")
print(df.dtypes)

print("-" * 50)

# ==========================================
# 11. New Column + Calculation
# ==========================================

df["Bonus_Marks"] = df["Student_Marks"] + 5

print("New Column Added")
print(df)

print("-" * 50)

# ==========================================
# Export CSV
# ==========================================

df.to_csv("updated_student_marks.csv", index=False)

print("New CSV Exported Successfully")

print("-" * 50)

# ==========================================
# 12. Merge DataFrames
# ==========================================

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Yash", "Rahul", "Aman"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "City": ["Nagpur", "Pune", "Mumbai"]
})

merged_df = pd.merge(df1, df2, on="ID")

print("Merged DataFrame")
print(merged_df)

print("-" * 50)

# ==========================================
# Concat DataFrames
# ==========================================

df3 = pd.DataFrame({
    "ID": [4, 5],
    "Name": ["Karan", "Rohit"]
})

concat_df = pd.concat([df1, df3], ignore_index=True)

print("Concatenated DataFrame")
print(concat_df)

print("-" * 50)

print("PANDAS REVISION COMPLETED ✅")
```
