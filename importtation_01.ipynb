```python
import pandas as pd

# Pandas me Series ek 1-dimensional data structure hota hai.
# Ye ek single column ki tarah data store karta hai.
```

```python
# Series create karna
names = pd.Series(["Rahul", "Priya"])

# Series print karna
print("Series:")
print(names)
```

```python
# DataFrame create karne ke liye dictionary use kar rahe hain.
data = {
    "Name": ["Rahul", "Priya"],
    "Age": [25, 23]
}

# DataFrame 2-dimensional data structure hota hai
# jisme rows aur columns dono hote hain.
df = pd.DataFrame(data)

# DataFrame print karna
print("\nDataFrame:")
print(df)
```
